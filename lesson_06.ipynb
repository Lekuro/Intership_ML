```python
with open("dataset.txt")as f:
  data=f.readlines()
xx=[]
fx=[]
f2x=[]
for line in data:
  x,y=line.split()
  try:
    xx.append(int(x))
    fx.append(float(y))
    f2x.append(2*float(y))
  except Exception as e:
    print(e)
print('xx',xx)
print('fx',fx)
print('f2x',f2x)
```

```python
def annot_max(x,y, ax=None):
    xmax = x[np.argmax(y)]
    ymax = max(y)
    text= "x={:.0f}, y={:.3f}".format(xmax, ymax)
    if not ax:
        ax=plt.gca()
    bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
    arrowprops=dict(arrowstyle="->",connectionstyle="angle,angleA=0,angleB=60")
    kw = dict(xycoords='data',textcoords="axes fraction",
              arrowprops=arrowprops, bbox=bbox_props, ha="right", va="top")
    ax.annotate(text, xy=(xmax, ymax), xytext=(0.94,0.96), **kw)
```

```python
plt.plot(xx,fx,'-g',label='f(x)')
plt.plot(xx,f2x,'--m', label='f(2*x)')
plt.legend(loc='lower left')
plt.xlabel('x', fontsize=15, color='blue')
plt.ylabel('f',fontsize=15,color='blue')
annot_max(xx,fx)
# plt.annotate('max',xy=(x_max,y_max), xycoords='data', xytext=(0,10),toxtxoords='data',arrowporps=dict(facecolor='m'))
plt.show()
```

```python
plt.step(xx,f2x,'m--o',where='mid')
plt.step(xx,fx,'g-o',where='pre')
annot_max(xx,fx)
plt.show()
```

```python
plt.stem(xx,fx, 'g-')
plt.stem(xx,f2x,'m:',markerfmt='*')
annot_max(xx,fx)
plt.show()
```

```python
plt.bar(xx,fx)
annot_max(xx,fx)
plt.show
```
