```python
genre_unique=frame['Genre'].unique()
genre_unique
```

```python
genre_collections=dict(tuple(frame.groupby('Genre')))
genre_collections
```

```python
genre_collections['Action']
```

```python
genre_collections['Adventure']
```

```python
genre_collections['Fighting'].sort_values('Year', ascending=True)
```

```python
for key,value in genre_collections.items():
  genre_collections[key]=value.sort_values('Year',ascending=True)
```

```python
genre_collections['Misc']
```

```python
genre_collections['Platform']
```

```python
genre_collections['Puzzle'].reset_index()
```

```python
genre_collections['Puzzle']
```

```python
for key,value in genre_collections.items():
  genre_collections[key]=value.reset_index()
```

```python
genre_collections['Racing']
```

```python
for key,value in genre_collections.items():
  value.to_csv(f'{key}.csv', encoding='utf-8')
```
